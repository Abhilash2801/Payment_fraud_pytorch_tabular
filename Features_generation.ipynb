```python
def haversine(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2.0)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    km = 6371 * c  # Radius of Earth in kilometers. Use 3956 for miles. Output distance in km.
    return km


# Apply the Haversine function to the DataFrame
df['distance'] = df.apply(lambda row: haversine(row['lat'], row['long'], row['merch_lat'], row['merch_long']), axis=1)

# Print the DataFrame to check the new featureb
```

```python
import pandas as pd
import numpy as np

# Define the new distance calculation function using the provided formula
def distance_using_acos(lat1, lon1, lat2, lon2):
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # Calculate the distance using the ACOS formula
    distance = np.arccos(
        np.sin(lat1) * np.sin(lat2) + 
        np.cos(lat1) * np.cos(lat2) * np.cos(lon2 - lon1)
    ) * 6371  # Radius of the Earth in kilometers
    
    return distance

# Assuming `df` is your dataframe
df['distance'] = df.apply(lambda row: distance_using_acos(row['lat'], row['long'], row['merch_lat'], row['merch_long']), axis=1)
```

```python
current_date = datetime.now().date()

# Define a function to calculate age
def calculate_age(dob, current_date):
    return current_date.year - dob.year - ((current_date.month, current_date.day) < (dob.month, dob.day))

# Apply the function to calculate the age for each entry
df['age'] = df['dob'].apply(lambda dob: calculate_age(dob, current_date))
```
